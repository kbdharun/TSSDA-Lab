```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
url = "https://www.ncei.noaa.gov/pub/data/swdi/stormevents/csvfiles/StormEvents_details-ftp_v1.0_d2010_c20220425.csv.gz"
df = pd.read_csv(url, compression='gzip')
# Filter data for tornadoes, hurricanes, and floods
tornado_df = df[df['EVENT_TYPE'] == 'Tornado']
hurricane_df = df[df['EVENT_TYPE'] == 'Hurricane (Typhoon)']
flood_df = df[df['EVENT_TYPE'] == 'Flood']
```

```python
# Convert 'BEGIN_DATE_TIME' to datetime
tornado_df['BEGIN_DATE_TIME'] = pd.to_datetime(tornado_df['BEGIN_DATE_TIME'])
hurricane_df['BEGIN_DATE_TIME'] = pd.to_datetime(hurricane_df['BEGIN_DATE_TIME'])
flood_df['BEGIN_DATE_TIME'] = pd.to_datetime(flood_df['BEGIN_DATE_TIME'])
```

```python
# Convert 'BEGIN_DATE_TIME' to datetime
tornado_df['BEGIN_DATE_TIME'] = pd.to_datetime(tornado_df['BEGIN_DATE_TIME'])
hurricane_df['BEGIN_DATE_TIME'] = pd.to_datetime(hurricane_df['BEGIN_DATE_TIME'])
flood_df['BEGIN_DATE_TIME'] = pd.to_datetime(flood_df['BEGIN_DATE_TIME'])
```

```python
# Plotting
plt.figure(figsize=(12, 6))

# Plot tornado events
sns.lineplot(data=tornado_df.groupby(tornado_df['BEGIN_DATE_TIME'].dt.date).size(), label='Tornado')
```

```python
# Check if the DataFrame is empty
print(hurricane_df.empty)

# Check if the 'BEGIN_DATE_TIME' column exists and is not empty
print('BEGIN_DATE_TIME' in hurricane_df.columns and not hurricane_df['BEGIN_DATE_TIME'].empty)

# Check the first few values of the 'BEGIN_DATE_TIME' column
print(hurricane_df['BEGIN_DATE_TIME'].head())
```

```python
# Plot hurricane events
sns.lineplot(data=hurricane_df.groupby(hurricane_df['BEGIN_DATE_TIME'].dt.date).size(), label='Hurricane')
```

```python
# Plot flood events
sns.lineplot(data=flood_df.groupby(flood_df['BEGIN_DATE_TIME'].dt.date).size(), label='Flood')

plt.title('Extreme Weather Events Over Time')
plt.xlabel('Date')
plt.ylabel('Event Count')
plt.legend(title='Event Type')
plt.show()
```
